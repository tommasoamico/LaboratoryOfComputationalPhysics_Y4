2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
# Outer method
outer_one = np.outer(u, v)
# Nested list method
outer_two_list = [i*j for i in u for j in v]
outer_two_final = np.reshape(np.array(outer_two_list), (len(u), len(v)))
# Broadcasting method
outer_three1 = np.ones((len(u), len(v)), dtype=int) * u.reshape(-1, 1) * v
print(" Outer product with numpy outer method: \n", outer_one)
print("\n Outer product with list comprehension: \n", outer_two_final)
print("\n Outer product with broadcasting: \n", outer_three1)
````

5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anto-diagonal (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagnoal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
init_matrix = np.ones((10, 10))
second_matrix = init_matrix * np.arange(1, 11)
multiplication_table = second_matrix * np.arange(1, 11).reshape(-1, 1)
print(" The multiplication table is:\n", multiplication_table)
trace = np.trace(multiplication_table)
print("\n Trace of multiplication table with np.trace method: \n", trace)
trace2 = reduce(lambda x, y: x+y, [multiplication_table[i, i] for i in range(np.shape(multiplication_table)[0])])
print("\n Trace of multiplication table with lambda function: \n", trace2)
anti_diagonal = np.fliplr(multiplication_table).diagonal()
print("\n Anti-diagonal with diagonal method: \n", anti_diagonal)
anti_diagonal2 = np.array([multiplication_table[i, 9-i] for i in np.arange(10)])
print("\n Anti-diagonal with list comprehension: \n", anti_diagonal2)
shifted_diagonal = np.diagonal(multiplication_table, offset=1)
print("\n Diagonal offset by one: \n", shifted_diagonal)
mask = np.ones(multiplication_table.shape[0], bool)
mask[9] = 0
crippled_matrix1 = multiplication_table[:,mask]
mask = np.ones(multiplication_table.shape[0], bool)
mask[0] = 0
crippled_matrix2 = multiplication_table[mask, :]
shift_diagonal1 = crippled_matrix2.diagonal()
print("\n Diagonal offset by one with mask: \n", shift_diagonal1)
````

6\. Use broadcasting to create a grid of distances

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles

  * Construct a 2D grid of distances among each city along Route 66

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
def primes(end):
    mask = np.ones(end, bool)
    mask[0], mask[1] = 0, 0
    for i in np.arange(2, end):
        if mask[i]:
            j = 2
            while i*j < end:
                mask[i*j] = 0
                j += 1
    list_of_numbers = np.array([i for i in np.arange(end)])
    list_of_primes = list_of_numbers[mask]
    return list_of_primes


print("The prime numbers between 0 and 99 are: \n", primes(100))
%timeit primes(100)
````

````python
%timeit primes(200)
%timeit primes(300)
%timeit primes(400)
%timeit primes(500)
%timeit primes(600)
%timeit primes(700)
%timeit primes(800)
%timeit primes(900)
%timeit primes(1000)
````

Eratosthenes

````python
import math


def eratosthenes_sieve(end):
    mask = np.ones(end, bool)
    mask[0], mask[1] = 0, 0
    for i in np.arange(2, math.floor(np.sqrt(end))):
        if mask[i]:
            j = 0
            while i**2 + j*i < end:
                mask[i**2 + j*i] = 0
                j += 1
    list_of_numbers = np.array([i for i in np.arange(end)])
    list_of_primes = list_of_numbers[mask]
    return list_of_primes


print("The prime numbers between 0 and 99 are: \n", eratosthenes_sieve(99))
%timeit eratosthenes_sieve(100)
````

````python
%timeit eratosthenes_sieve(200)
%timeit eratosthenes_sieve(300)
%timeit eratosthenes_sieve(400)
%timeit eratosthenes_sieve(500)
%timeit eratosthenes_sieve(600)
%timeit eratosthenes_sieve(700)
%timeit eratosthenes_sieve(800)
%timeit eratosthenes_sieve(900)
%timeit eratosthenes_sieve(1000)
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a row and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
np.random.seed(123) # fix seed for reproducible result
````

````python
first_matrix = np.random.rand(1000, 200)
first_matrix[first_matrix < 0.5] = -1
first_matrix[first_matrix >= 0.5] = 1
print(" Sample of the matrix of single steps: \n", first_matrix[:10, :10])
distance_matrix = np.cumsum(first_matrix, axis=1)
print("\n Sample of the matrix tracking the distance covered: \n", distance_matrix[:10, :10])
squared_distance_matrix = distance_matrix**2
mean_per_step = np.sqrt(np.mean(squared_distance_matrix, axis=0))
x_axis = np.array(np.arange(200))
y_axis = mean_per_step
plt.title('Random walk plot')
plt.xlabel('Steps')
plt.ylabel('Average distances')
plt.plot(x_axis, y_axis)
plt.show()
````
