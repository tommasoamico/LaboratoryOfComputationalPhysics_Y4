1. The MickeyMouse problem

a) Write a program that prints the numbers from 1 to 100. But for multiples of three print Mickey instead of the number and for the multiples of five print Mouse. For numbers which are multiples of both three and five print MickeyMouse

b) Put the result in a list and substitute Mickey with Donald and Mouse with Duck

```python
for number in range(1, 101):
    # the order of the if statements matter
    if number % 3 == 0 and number % 5 == 0:
        print("MickeyMouse")
        continue
    elif number % 3 == 0:
        print("Mickey")
        continue
    elif number % 5 == 0:
        print("Mouse")
        continue    
    else:
        print(number)
```

2\. The swap function

Write a function that swap the values of two input variables x and y (whatever the type). Try to do that also without a temporary variable

```python
x = input("Type the first variable: ")
y = input("Type the second variable: ")
def swap(a, b):
    a, b = b, a
    return a, b 


x, y = swap(x, y)
print(x)
print(y)
```

3\. Computing the distance

Write a function that calculates and returns the euclidean distance between two points *u* and *v*, where *u* and *v* are both 2-tuples *(x,y)*. For example, if *u=(3,0)* and *v=(0,4)*, the function should return 5

```python
import math
from collections import namedtuple

try:
    point_1_x = int(input("Type the x of the first point: "))
    point_1_y = int(input("Type the y of the first point: "))
    point_2_x = int(input("Type the x of the second point: "))
    point_2_y = int(input("Type the y of the second point: "))
    Point = namedtuple("Point", "x, y")
    point_1 = Point(point_1_x, point_1_y)
    point_2 = Point(point_2_x, point_2_y)
    def euclidean_distance(first_point, second_point):
        distance = math.sqrt((first_point.x-second_point.x)**2+(first_point.y-second_point.y)**2)
        return distance

    print(euclidean_distance(point_1, point_2))
except ValueError:
    print("You must type a number")
```

6\. Combination of functions

Write two functions - one that returns the square of a number, and one that returns the cube. Now write a third function that returns the number raised to the 6th power using the two previous functions.

```python
number = int(input("Type your number: "))
def square(number1):
    return number1*number1


def cube(number2):
    return number2**3


def sixth_power(number3):
    return cube(square(number3))


print(f"The second power of {number} is {square(number)}")
print(f"The third power of {number} is {cube(number)}")
print(f"The sixth power of {number} is {sixth_power(number)}")
```

9\. Normalization

Write a function that takes a tuple of numbers and returns it with the entries normalized to one

```python
import math

def normalization(tuples):
    prefactor = 0
    for number in tuples:
        prefactor += number**2
    #Here we change the tuple in a list to modify it
    mylist = list(tuples)
    for i in range(0, len(mylist)):
        mylist[i] = mylist[i]/math.sqrt(prefactor)
    #Here we come back to the tuple as requested
    mytuple = tuple(mylist)
    return(mytuple)

print(normalization((3, 4, 5)))
```
